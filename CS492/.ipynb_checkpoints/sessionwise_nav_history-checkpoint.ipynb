# Get maxtime

```python
import os
import glob
import json
import csv
import re


curdir=os.getcwd()

targetvideolist='targetvideos.csv'
with open(targetvideolist,'r',encoding='utf-8') as rf:
    reader=csv.reader(rf)
    targetvideos=[]
    next(reader)
    for row in reader:
        targetvideos.append(row[0])
    rf.close()

    
data=json.load(open('video_targetdata_session.json','r',encoding='utf-8'))

maxtimes=[]

for vid in targetvideos:
    for vobj in data:
        if(vobj["video_id"]==vid):
            maxtime=0
            for s in vobj["sessions"]:
                for event in s["events"]:
                    loctime=0
                    try:
                        loctime=event["currentTime"]
                    except:
                        try: 
                            loctime=max(event["old_time"],event["new_time"])
                        except:
                            pass
                maxtime=max(maxtime, loctime)
    newobj={"video_id":vid, "max_time":float(maxtime)}
    maxtimes.append(newobj)

with open('video_maxtime.json','w',encoding='utf-8') as wf:
    eventwriter=csv.writer(wf,delimiter=',')
    json.dump(maxtimes,wf, sort_keys=False, indent=4)
    wf.close()

```

# Craete grid for timestamps

```python
# manage session-wise navigation history
import math

#generate grid
navdata=[]
for vid in targetvideos:
    for tobj in maxtimes:
        if(tobj["video_id"]==vid):
            maxtime=tobj["max_time"]
    for vobj in data:
        if(vobj["video_id"]==vid):
            newnavdata={}
            newnavdata["video_id"]=vid
            newnavdata["sessions"]=[]
            
            for s in vobj["sessions"]:
                navhist=[]
                for i in range(math.floor(maxtime)-1):
                    navhist.append({"timestamp":i, "seen":0, "speed": None})
                navhist.append({"timestamp":math.floor(maxtime)-1, "seen":0, "speed": None, "duration":0})
                snavhist={}
                snavhist["session"]=s["session"]
                snavhist["user_id"]=s["user_id"]
                snavhist["navhist"]=navhist
                newnavdata["sessions"].append(snavhist)
            navdata.append(newnavdata)

with open('video_navhist.json','w',encoding='utf-8') as wf:
    eventwriter=csv.writer(wf,delimiter=',')
    json.dump(navdata,wf, sort_keys=False, indent=4)
    wf.close()                

```

# Calculate navigation count and speed for each timestamps

```python
navdata=json.load(open('video_navhist.json','r',encoding='utf-8'))


def fill_navmode(navhist, start, end, speed):
    left=min(start,end)
    right=max(start,end)
    for hist in navhist:
        if(hist["timestamp"] in range(left, right)):
            hist["seen"]=hist["seen"]+1
            if(hist["speed"]):
                hist["speed"]=min(hist["speed"], speed)
            else:
                hist["speed"]=speed
    return navhist

def get_curtime(event):
    curtime=event["currentTime"]
    if isinstance(curtime, float):
        curtime=math.floor(curtime)
    else:
        curtime=0
    return curtime

def fill_firstspeed(navhist,curtime, oldspeed):
    for hist in navhist:
        if(hist["timestamp"] in range(curtime)):
            if(hist["seen"]>0):
                if(hist["speed"]):
                    hist["speed"]=min(hist["speed"],oldspeed)
                else:
                    hist["speed"]=oldspeed
    return navhist
    
        
for vid in targetvideos:
    for vobj in data:
        if(vobj["video_id"]==vid):
            for s in vobj["sessions"]:
                # get navhist
                for hobj in navdata:
                    if(hobj["video_id"]==vid):
                        hsessions=hobj["sessions"]
                        for hs in hsessions:
                            if(hs["session"]==s["session"]):
                                navhist=hs["navhist"]
                navmode=0
                prevtime=0
                currentspeed=1
                firstchange=1
                for event in s["events"]:
                    et=event["event_type"]
                    if et=="play_video":
                        prevtime=get_curtime(event)   
                        navmode=1
                    elif et=="pause_video":
                        if navmode==1:
                            navhist=fill_navmode(navhist, prevtime, get_curtime(event), currentspeed)
                        prevtime=get_curtime(event)
                        navmode=0
                    elif et=="stop_video":
                        if navmode==1:
                            navhist=fill_navmode(navhist, prevtime,get_curtime(event),currentspeed)
                        prevtime=get_curtime(event)
                        navmode=0
                    elif et=="seek_video":
                        if navmode==1:
                            navhist=fill_navmode(navhist, prevtime, math.floor(event["old_time"]),currentspeed)
                        prevtime=math.floor(event["new_time"] )                     
                    elif et=="speed_change_video":
                        if navmode==1:
                            if(firstchange):
                                navhist=fill_firstspeed(navhist,get_curtime(event),event["old_speed"])
                                firstchange=0
                            else:
                                if currentspeed != event["old_speed"]:
                                    print(currentspeed, event["old_speed"], s["session"])
                                navhist=fill_navmode(navhist,prevtime,get_curtime(event),currentspeed)
                        currentspeed=event["new_speed"]
                        prevtime=get_curtime(event)
                    else:
                        continue
                
                for hobj in navdata:
                    if(hobj["video_id"]==vid):
                        hsessions=hobj["sessions"]
                        for hs in hsessions:
                            if(hs["session"]==s["session"]):
                                hs["navhist"]=navhist


with open('video_navhist_countspeed.json','w',encoding='utf-8') as wf:
    eventwriter=csv.writer(wf,delimiter=',')
    json.dump(navdata,wf, sort_keys=False, indent=4)
    wf.close()
```

# Calculate pause duration for each bin

```python
# define bin width 
bw=60

def get_bintime(curtime):
    bintime=math.floor(curtime/bw)*bw
    return bintime
    
pausedata=json.load(open('video_targetdata_session_with_duration.json','r',encoding='utf-8'))
for vobj in pausedata:
    vid=vobj["video_id"]
    for s in vobj["sessions"]:
        sid=s["session"]
        for e in s["events"]:
            if (e["event_type"]=="pause_video"):
                try:
                    duration = e["duration"]
                    curtime=get_curtime(e)
                    bintime=get_bintime(curtime)
                    for nvobj in navdata:
                        if(nvobj["video_id"]==vid):
                            nsessions=nvobj["sessions"]
                            for ns in nsessions:
                                if(ns["session"]==sid):
                                    navs=ns["navhist"]
                                    for hist in navs:
                                        if(hist["timestamp"]==bintime):
                                            try:
                                                curduration=hist["duration"]
                                                hist["duration"]=curduration+duration
                                            except:
                                                hist["duration"]=duration
                except KeyError:
                    # Key is not present
                    pass

with open('video_navhist_countspeedpause.json','w',encoding='utf-8') as wf:
    eventwriter=csv.writer(wf,delimiter=',')
    json.dump(navdata,wf, sort_keys=False, indent=4)
    wf.close()                  
```
